### Data preparation
1. Droping irrelevant columns based on requirements
2. Renaming the right colunms

```python
  """Creating a new copy of BasicCompanyData by selecting the required columns CompanyCategory, 
  CompanyStatus and Accounts.AccountCategory
  """
    
BasicCompany_Data = BasicCompanyData[['CompanyCategory','CompanyStatus','Accounts.AccountCategory']].copy()
```

```python
"""showing the available columns in BasicCompany_Data."""

display('BasicCompanyData columns:',BasicCompany_Data.columns)
```

```python
"""Verifing the newly copied dataset"""

display(BasicCompany_Data.head())
```

```python
""" Understanding CompanyCategory field"""

print(BasicCompany_Data['CompanyCategory'].value_counts())
```

```python
""" Filtering Private Limited Company fom newBCD"""

BasicCompany_Data_PLC = BasicCompany_Data[BasicCompany_Data['CompanyCategory'] == 'Private Limited Company']
```

```python
"""Checking the unique categories in Accounts.AccountCategory"""

print('Unique Account Category:',BasicCompany_Data_PLC['CompanyCategory'].nunique()) 
```

```python
""" Filtering 'Active' fom CompanyStatus"""

BasicCompany_Data_Active = BasicCompany_Data_PLC[BasicCompany_Data_PLC['CompanyStatus'] == 'Active']
```

```python
"""Checking the unique Active in CompanyStatus"""

print('Unique CompanyStatus:',BasicCompany_Data_Active['CompanyStatus'].nunique()) 
```

```python
"""Verifying the various categories in Accounts.AccountCategory"""

print(BasicCompany_Data_Active['Accounts.AccountCategory'].value_counts())
```

```python
""" Renaming Accounts.AccountCategory to AccountCategory"""

BasicCompany_Data_Active.rename(columns={'Accounts.AccountCategory': 'AccountCategory'}, inplace=True)
```

```python
""" Filtering 'Small' and 'Medium' fom CompanyStatus"""

BasicCompany_Data_SM = BasicCompany_Data_Active[(BasicCompany_Data_Active['AccountCategory'] == 'SMALL') | (BasicCompany_Data_Active['AccountCategory'] == 'MEDIUM')]
```

```python
"""Checking the unique categories in AccountCategory"""

print('Unique Account Category:',BasicCompany_Data_SM['AccountCategory'].nunique()) 
```

```python
""" Verifying that Accounts Categories contains the filtered; ‘SMALL’ and ‘MEDIUM’"""

print(BasicCompany_Data_SM['AccountCategory'].value_counts())
```

```python
""" Resetting the index using reset_index()"""

processed_BasicCompanyData = BasicCompany_Data_SM.reset_index(drop=True)
display(processed_BasicCompanyData.head())
display(processed_BasicCompanyData.tail())
```

```python
""" Checking the processed_BasicCompanyData shape"""
""" 2891 rows and 3 columns """

print('processedData :',processed_BasicCompanyData.shape)
```

```python
"""Checking the info of the newly created processedData"""
"""No missing entries identified."""

print(processed_BasicCompanyData.info())
```

```python
"""Saving processed_BasicCompanyData as output.csv"""

processed_BasicCompanyData.to_csv('filtered_BasicCompanyData.csv', index=False)
```
